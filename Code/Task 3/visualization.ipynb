```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from PIL import Image
import json


def draw_rectangle(img,bounding_box):
  starting_point = (int(bounding_box["x"]-bounding_box["width"]/2),int(bounding_box["y"]-bounding_box["height"]/2))
  ending_point = (int(bounding_box["x"]+bounding_box["width"]/2),int(bounding_box["y"]+bounding_box["height"]/2))
  cv2.rectangle(img,starting_point,ending_point,(255,0,0),thickness=2)
  return img

def draw_point(img,point):
  # Draw a point at the coordinates (x, y)
  circle_radius = 1
  cv2.circle(img, (int(point["x"]), int(point["y"])), circle_radius, (0, 255, 0), thickness=2)
  return img

def visualize_annotation(frame,annotation):
  # Load the image
  img = cv2.imread(frame)
  img = np.asarray(Image.open(frame))

  # Get image dimensions
  height, width = img.shape[:2]

  # Draw a rectangle in the bounding box coordinates
  # bounding_box = annotation["annotation"][0]
  # img = draw_rectangle(img,bounding_box)

  # Draw points
  for point in annotation["annotations"]:
    if "class" in point.keys():
      img = draw_point(img,point)
    elif "mouse1" or "mouse2" in point.keys():
      img = draw_rectangle(img,point)

  # Show the modified image
  cv2_imshow(img)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
```

```python
def visualize_json(json_file_path,images_file_path):
  with open(json_file_path, 'r') as f:
    # Load the JSON data into a dictionary
    annotations = json.load(f)

  for annotation in annotations:
    img = annotation["filename"]
    img = img.replace("png","jpg")
    visualize_annotation(images_file_path+"/"+img,annotation)
```
